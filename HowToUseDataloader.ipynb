# How to use dataloader

```python
import pandas as pd
```

```python
def make_df(loader):
    data = loader()
    columns = data.get("feature_names")
    df = pd.DataFrame(data.data, columns=columns)
    df["target"] = pd.Series(data.target)
    return df
```

```python
from sklearn.datasets import load_digits
```

```python
import dataloader
```

```python
from sqlalchemy import create_engine
engine = create_engine("teradata://dbc:dbc@192.168.43.24:1025/tdwork")
```

```python
df_digits= make_df(load_digits)
```

```python
df_digits= make_df(load_digits)
dataloader.tdload_df(df_digits, engine, tablename="tdload_digits", overwrite=True,
          ssh_ip="192.168.43.24", ssh_username="root", ssh_password="root",
                    verbose=False, saveIndex=True, indexList=["index"])
```
